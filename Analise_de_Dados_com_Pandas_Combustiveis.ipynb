```python
#Aqui criamos um novo dataframe apenas com as colunas desejadas 
ca_df = comb_df[['Revenda','Municipio','Produto','Valor de Venda']]
display(ca_df)
```

```python
#Locaizando preÃ§os de etanol em INDAITUBA
# Sort

etanol_inda_df = ca_df.loc[(ca_df['Produto'] == 'ETANOL') & (ca_df ['Municipio']== 'INDAIATUBA')]
etanol_inda_df.sort_values(by='Valor de Venda', inplace = True)
display(etanol_inda_df)
```

```python
#Qual a media de preÃ§os dos combustiveis gasolina e gasolina aditivada do bairro Bela vista em sÃ£o paulo
#funcao isin

#comb_sp = comb_df.loc[(comb_df['Bairro'] == 'MOOCA') & (comb_df['Municipio'] == 'SAO PAULO') & ((comb_df['Produto'] == 'GASOLINA') | (comb_df['Produto'] == 'GASOLINA ADITIVADA'))]

comb_sp = comb_df.loc[(comb_df['Bairro'] == 'MOOCA') & 
                      (comb_df['Municipio'] == 'SAO PAULO') & 
                      (comb_df['Produto'].isin(['GASOLINA', 'GASOLINA ADITIVADA']))]

comb_sp = comb_sp [['Revenda','Municipio','Bairro','Produto','Valor de Venda']]
comb_sp.sort_values(by='Valor de Venda', inplace = True)
display(comb_sp.mean())
display(comb_sp)
```

```python
# Media de valor geral por tipo de combustivel
# Round para casas decimais

media_comb_df = ca_df[['Produto', 'Valor de Venda']].groupby(by='Produto').mean().round(2)
display(media_comb_df)
```
